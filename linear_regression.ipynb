```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
df = pd.read_csv("land_prices.csv")

x= df[['Land Area(acres)', 'distance(m)']]
y = df['price(M)']

model = LinearRegression()
model.fit(x,y)

try:
    user_size = float(input("Enter the land size in acres to predict the price: "))
    user_dist = float(input("Enter the distance from the road in metres: "))
    new_data= pd.DataFrame([[user_size, user_dist]],
                           columns=['Land Area(acres)', 'distance(m)'])
    prediction = model.predict(new_data)
    print("\n" + "="*30)
    print(f"PREDICTION REPORT")
    print(f"Area: {user_size} Acres")
    print(f"Distance: {user_dist} KM from road")
    print(f"Estimated Price: KES {prediction.item():.2f} Million")
    print("="*30)

except ValueError:
    print("Invalid input! Please enter a number (e.g., 5.5).")
```
