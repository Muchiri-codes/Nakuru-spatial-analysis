```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

df = pd.read_csv("land_prices.csv")

x= df[['Land Area(acres)']]
y = df['price(M)']

model = LinearRegression()
model.fit(x,y)

try:
    user_size = float(input("Enter the land size in acres to predict the price: "))
    new_data= pd.DataFrame([[user_size]], columns=['Land Area(acres)'])
    prediction = model.predict(new_data)
    print("_" *30)
    print(f"For {user_size} acres")
    print(f"The predicted price for 10 acres is: KES {prediction.item():.2f} Million")
    print("_" *30)

except ValueError:
    print("Invalid input! Please enter a number (e.g., 5.5).")
plt.scatter(x, y, color="blue")
plt.plot(x, model.predict(x), color='red')
plt.xlabel("Area in Acres")
plt.ylabel("Price in Millions")
plt.title("Land price using linear regression")
plt.show()
```
