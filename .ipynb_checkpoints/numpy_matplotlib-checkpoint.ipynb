```python
import numpy as np
elevation = np.array([[100.32, 100.13, 98.99, 100.11, 100.00],
                      [100.01, 100.22, 99.91, 100.33, 100.20],
                      [100.25, 100.37, 100.42, 100.29, 100.26],
                      [101.34, 102.43, 103.21, 102.43, 103.37],
                      [101.65, 104.32, 103.32, 104.00, 104.32]
                      ])
```

# BASIC STATS

```python
mean_elv = np.mean(elevation)
max_elv = np.max(elevation)
min_elv = np.mean(elevation)
std_dev = np.std(elevation)


print("mean:", mean_elv)
print("max:", max_elv)
print("min:", min_elv)
print("standard deviation:", std_dev)
```

# FIND HIGH AREAS

```python
high_points = elevation > (mean_elv + std_dev)
print("High elevation points:\n", high_points)
```

# VISUALIZATION

```python
import matplotlib.pyplot as plt
plt.imshow(elevation, cmap='terrain')
plt.colorbar(label='Elevation (M)')
plt.show()
```

## USING DEM

```python
import numpy as np
import rasterio
import matplotlib.pyplot as plt

dem_data = "Nakuru_dem_mosaic.tif"
with rasterio.open(dem_data) as dataset:
    dem = dataset.read(1)
    if dataset.nodata is not None:
        dem = np.where(np.isclose(dem, dataset.nodata), np.nan, dem)

print("DEM Shape:",  dem.shape)
print("mean DEM value:", np.nanmean(dem))
print("max DEM value:", np.nanmax(dem))
print("min DEM value:", np.nanmin(dem))
print("standard deviation of DEM value:", np.nanstd(dem))
print("No data values:", dataset.nodata)
```

# Quick visualization of the dem data

```python
plt.figure(figsize=(10,8))
cmap = plt.cm.terrain.copy()
cmap.set_bad(color='grey')
plt.imshow(dem, cmap= cmap)
plt.colorbar(label='Elevation(M)')
plt.title("NAKURU DEM VISUALIZATION")
plt.show
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

df = pd.read_csv("Topo parcel 4.csv")
#only use these 3 columns to make your decision
X_scaled= df[['lon', 'lat', 'elev']]
scaler = StandardScaler()

model = KMeans(n_clusters=3, random_state=42)
df['zone_cluster'] = model.fit_predict(X_scaled)

#visualization
plt.figure(figsize=(10, 8))
plt.scatter(df['lat'], df['lon'], c=df['zone_cluster'], cmap='viridis', s=50)
plt.title("AUTOMATIC LAND ZONING BASED ON TOPO DATA")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.colorbar(label="ZONE ID")
plt.show()
```
